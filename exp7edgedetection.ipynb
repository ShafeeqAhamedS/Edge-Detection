```python
import cv2
from cv2 import COLOR_BGR2GRAY
import matplotlib.pyplot as plt
```

```python
input_img = cv2.imread("mikasa.jpg",1)
gray = cv2.cvtColor(input_img,COLOR_BGR2GRAY)
cv2.imshow("Input",input_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.imshow("Gray",gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# remove noise
img = cv2.GaussianBlur(gray,(3,3),0)
cv2.imshow("Window",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=9)
sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=13)
sobelxy =cv2.Sobel(img,cv2.CV_64F,1,1,ksize=31)

cv2.imwrite("sobelx.png",sobelx)
cv2.imwrite("sobely.png",sobely)
cv2.imwrite("sobelxy.png",sobelxy)
```

```python
plt.imshow(img,cmap = 'gray')
plt.title('Original')
plt.xticks([]), plt.yticks([])

```

```python
plt.imshow(sobelx,cmap = 'gray')
plt.title('sobelx')
plt.xticks([]), plt.yticks([])
```

```python
plt.imshow(sobely,cmap = 'gray')
plt.title('sobely')
plt.xticks([]), plt.yticks([])
```

```python
plt.imshow(sobelxy,cmap = 'gray')
plt.title('sobelxy')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
plt.figure(1)

plt.subplot(2,2,1)
plt.imshow(img,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])

plt.subplot(2,2,2)
plt.imshow(sobelx,cmap = 'gray')
plt.title('sobelx')

plt.xticks([]), plt.yticks([])

plt.subplot(2,2,3)
plt.imshow(sobely,cmap = 'gray')
plt.title('sobely')
plt.xticks([]), plt.yticks([])

plt.subplot(2,2,4)
plt.imshow(sobelxy,cmap = 'gray')
plt.title('sobelxy')
plt.xticks([]), plt.yticks([])

plt.show()
```

```python
laplacian = cv2.Laplacian(img,cv2.CV_64F)
plt.imshow(laplacian,cmap = 'gray')
plt.title('laplacian')
plt.axis("off")
plt.show()
```

```python
canny_edges = cv2.Canny(img, 120, 150)
plt.imshow(canny_edges,cmap = 'gray')
plt.title('canny_edges')
plt.axis("off")
plt.show()
cv2.imshow("Canny",canny_edges)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
